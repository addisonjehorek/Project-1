# Task 3

```python
# pm2.5 + pm10.0
air_data3 = air_data




#step 1: define a funcion that assigns temperature levels 

def assign_temperature(temperature): 
    if temperature < 32: 
        return "Below Freezing" 
    elif 32<= temperature <= 50: 
        return "Cool"
    elif 51 <= temperature <= 70: 
        return "Warm" 
    else: 
        return "Hot"

def assign_humidity(humidity): 
    if humidity < 50: 
        return "Low" 
    elif 51<= humidity <= 80: 
        return "High"
    else: 
        return "Very High" 

air_data3["temperature_level"] = air_data3["temperature"].apply(assign_temperature)
air_data3["humidity_level"] = air_data3["humidity"].apply(assign_humidity)


```

```python
# pm2.5 + pm10.0
air_data3 = air_data
stat_summary = air_data3.groupby(['humidity_level', 'temperature']).agg({
    'pm2.5_atm': ['max','min','mean','median'],
    'pm10.0_atm': ['max','min','mean','median']

    
})
#stat_summary

#Define pivot table
pivot_stats = pd.pivot_table(
    air_data3,
    index = ['humidity_level', 'temperature_level'],
    values = ['pm2.5_atm','pm10.0_atm'],
    aggfunc = ['max','min','mean','median']
)

#Call 
pivot_stats
```
